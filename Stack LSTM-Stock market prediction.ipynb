```python
#importing scaler package
from sklearn.preprocessing import MinMaxScaler
#scale down from 0 to 1
scaler=MinMaxScaler(feature_range=(0,1))


df1=scaler.fit_transform(np.array(df1).reshape(-1,1))
```

```python
##splitting dataset into train and test split

# timeseries data: is the data wrt time and ordered in time
#suppose day1 my stock prce was 150, dy 2: 170 day 3: 145 etc
# day2 is dependant on day 1, day 3 on day 2 n 1 etc


#to spilt these data
#make sure all the data is arranged according to time/date.take fisrt 70% data as train and rest as test 
# as data is dependant on previous data

training_size=int(len(df1)*0.65) #65%. of total lenght of my df as. training data

test_size=len(df1)-training_size # rest of data: 100-65

#row: 0th to 870th index as training. all columns
train_data=df1[0:training_size,:]
# from end of training_size till end of te df1 and all columns
test_data=df1[training_size:len(df1),:1]
```

```python
### Step 2: Data Preprocessing

#Consider my input data as: 120,130, 125, 140, 134, 150, 160, 190, 154, 170
#Based on time, I have spilt data as forst 60% as training and rest as test

#Consider my training data as: 120,130, 125, 140, 134, 150 
#test data: 160, 190, 154, 170

### How to split in X_train and y_train

#As each data is dependant on previous entries. 

#Consider timesteps as 3, where I will take 1st 3 etries as 1st 2 features f1, f2, and f3. 
#then 4th entry as o/p. In next step, I will move a step right.. 


#X_train             y_train
#f1     f2     f3     op

#120    130    125    140

#130    125     140   134


#where features f1, f2 and f3 are X_train and y_train is op 
#This data is used to train 


#Repeating for test data as well 

#y_train                y_test

#f1        f2     f3    op

#160       190   154   170

#190       154   170   160

#This data is used to test and will be compared for accuracy 
```

```python
#data preprocessing 


import numpy
# convert an array of values into a dataset matrix
def create_dataset(dataset, time_step=1): 
	dataX, dataY = [], []
	for i in range(len(dataset)-time_step-1):
		a = dataset[i:(i+time_step), 0]#i=0, 0,1,2,3-----99,100, time_step is 100 so 0th to 99th will be in X_train 
		dataX.append(a) # 100th element in y_train
		dataY.append(dataset[i + time_step, 0])# 101st element is dataY
	return numpy.array(dataX), numpy.array(dataY)
```

```python
# this is the next step for time_step= 100. timestep is hyper parameter
```

```python
# reshape into X=t,t+1,t+2,t+3 and Y=t+4
time_step = 100
X_train, y_train = create_dataset(train_data, time_step)
X_test, ytest = create_dataset(test_data, time_step)
```

```python
print(X_train.shape), print(y_train.shape)
```

```python
print(X_test.shape), print(ytest.shape)
```

```python
# reshape input to be [samples, time steps, features] which is required for LSTM
X_train =X_train.reshape(X_train.shape[0],X_train.shape[1] , 1)
X_test = X_test.reshape(X_test.shape[0],X_test.shape[1] , 1)
                        
                        
#reshaping into 3D, now X_train and X_test is in 2D, reshape it adding 1
# because the input will be 2nd and 3rd X_train.shape[1] , 1
                        
#The input of the LSTM is always is a 3D array. (batch_size, time_steps, seq_len) . 
#The output of the LSTM could be a 2D array or 3D array depending upon the return_sequences argument.
```

```python
### step3 : Create the Stacked LSTM model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
```

```python
model=Sequential()
#first lstm layer with 50 neurons. input shape is 100,1 because it is the shape of X_train.shape[1] , 1
model.add(LSTM(50,return_sequences=True,input_shape=(100,1)))
#as this is stack lstm, add one more layer 
model.add(LSTM(50,return_sequences=True))
#3rd lstm
model.add(LSTM(50))
#op layer 
model.add(Dense(1))
# calculate loss and reduce loss using adam's optimizer
model.compile(loss='mean_squared_error',optimizer='adam')
```

```python
model.summary()
```

```python
model.summary()
```

```python
model.fit(X_train,y_train,validation_data=(X_test,ytest),epochs=100,batch_size=64,verbose=1)
```

```python
#step 4: predict the test data and plot the output
import tensorflow as tf
```

```python
tf.__version__
```

```python
### Lets Do the prediction and check performance metrics
train_predict=model.predict(X_train)
test_predict=model.predict(X_test)
```

```python
##Transformback to original form
# we have scaled down btw 0 and 1 
#now we need to reverse that back 

train_predict=scaler.inverse_transform(train_predict)
test_predict=scaler.inverse_transform(test_predict)
```

```python
### Calculate RMSE performance metrics

#calculate mse btw y-train and train predict. and then calculate sqrt to cal rmse
import math
from sklearn.metrics import mean_squared_error
math.sqrt(mean_squared_error(y_train,train_predict))
```

```python
### Test Data RMSE
math.sqrt(mean_squared_error(ytest,test_predict))
```

```python
### Plotting 
# shift train predictions for plotting
look_back=100
trainPredictPlot = numpy.empty_like(df1)
trainPredictPlot[:, :] = np.nan
trainPredictPlot[look_back:len(train_predict)+look_back, :] = train_predict
# shift test predictions for plotting
testPredictPlot = numpy.empty_like(df1)
testPredictPlot[:, :] = numpy.nan
testPredictPlot[len(train_predict)+(look_back*2)+1:len(df1)-1, :] = test_predict
# plot baseline and predictions
plt.plot(scaler.inverse_transform(df1))
plt.plot(trainPredictPlot)
plt.plot(testPredictPlot)
plt.show()

#green is test data
#this is as similar as the graph we plotted for original data. so accuracy is good 
# in epochs we saw that the loss was 0.318 and ended with 3.4 to power -4. loss has reduced greatly
#this is not forecasting
```
